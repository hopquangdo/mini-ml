```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_metric_lines(models, metrics, scores, title='So sánh các mô hình'):
    scores = np.array(scores)
    plt.figure(figsize=(10, 6))

    for i, model in enumerate(models):
        plt.plot(metrics, scores[i], marker='o', label=model)

    plt.title(title)
    plt.xlabel("Tiêu chí")
    plt.ylabel("Giá trị")
    plt.grid(True)

    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    plt.tight_layout(rect=[0, 0, 0.85, 1])
    plt.show()



def plot_runtime_bar(models, runtimes, title="So sánh thời gian chạy (Runtime)"):
    plt.figure(figsize=(10, 5))
    plt.bar(models, runtimes, color='skyblue')
    plt.ylabel("Thời gian (giây)")
    plt.title(title)
    for i, rt in enumerate(runtimes):
        plt.text(i, rt + 0.05, f"{rt:.2f}", ha='center')
    plt.tight_layout()
    plt.show()
```

```python
models = ['Linear Regression', 'KNN', 'Decision Tree', 'SVR', 'KMeans']
metrics = ['MSE', 'MAE', 'R2']
scores = np.array([
    [0.5179331255246697, 0.5286283596582387, 0.6125511913966952],
    [0.0, 0.0, 1.0],
    [0.4843, 0.5063, 0.6377],
    [1.3685, 0.8715, -0.0237],
    [1.3354063010230042, 0.913632710677265, 0.0010262815135104741]
])
runtimes = [0, 0.0811, 0.0020, 15.2116, 0.07545638084411621]

plot_metric_lines(models, metrics, scores, title='So sánh các mô hình hồi quy')
plot_runtime_bar(models, runtimes)
```
